```python
def get_street_type(address):
    exclude_list = ['N', 'S', 'W', 'E']
    address_list = address.split(' ')
    street_type = address_list[-1]
    if street_type in exclude_list:
        street_type = address_list[-2]
    return street_type
```

```python
street_types = melb_df['Address'].apply(get_street_type)
display(street_types)
```

```python
def reduce_street_type_name(street_type: str) -> str:
    type_dict = {'Avenue': 'Av', 'Ave': 'Av', 'Boulevard': 'Bvd', 'Parade': 'Pde'}
    for long_type, short_type in type_dict.items():
        if street_type == long_type:
            street_type = short_type
    return street_type
```

```python
street_types = street_types.apply(reduce_street_type_name)
```

```python
import re
def get_year_release(arg):
    candidates = re.findall(r'\(\d{4}\)', arg)
    if len(candidates) > 0:
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        return None
```
