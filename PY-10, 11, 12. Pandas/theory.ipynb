```python
def get_street_type(address):
    exclude_list = ['N', 'S', 'W', 'E']
    address_list = address.split(' ')
    street_type = address_list[-1]
    if street_type in exclude_list:
        street_type = address_list[-2]
    return street_type
```

```python
street_types = melb_df['Address'].apply(get_street_type)
display(street_types)
```

```python
def reduce_street_type_name(street_type: str) -> str:
    type_dict = {'Avenue': 'Av', 'Ave': 'Av', 'Boulevard': 'Bvd', 'Parade': 'Pde'}
    for long_type, short_type in type_dict.items():
        if street_type == long_type:
            street_type = short_type
    return street_type
```

```python
street_types = street_types.apply(reduce_street_type_name)
```

```python
print(street_types.unique())
```

```python
display(street_types.value_counts())
```

```python
popular_types = street_types.value_counts().nlargest(10).index
print(popular_types)
```

```python
melb_df['StreetType'] = street_types.apply(lambda x: x if x in popular_types else 'other')
display(melb_df['StreetType'])
```

```python
count_unique_values = dict(melb_df.nunique())
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']

def convert_to_category(df:pd.DataFrame) -> pd.DataFrame:
    global count_unique_values
    global cols_to_exclude
    for col_name, n in count_unique_values.items():
        if n < 150 and col_name not in cols_to_exclude:
            df[col_name] = df[col_name].astype('category')
    return df

melb_df = convert_to_category(melb_df)
```
